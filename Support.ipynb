```python
# Wersja 02.11.2024

import napari
from skimage import io
import os
import numpy as np
from scipy.io import savemat
from scipy.io import loadmat
from scipy.ndimage import zoom
from numpy.fft import fft2, ifft2, fftshift, ifftshift
from scipy.ndimage import median_filter
import tifffile
import h5py
import time


# Wycięcie jedynie obaszru komórki
def masking_cell_area(image):
    # Wyciągnięcie macierzy zawierającej komórkę
    macierz_cell = image[:, 2, :, :]
    # Normalizacja pikseli do 1
    min_value_macierz_cell = np.min(macierz_cell) - 1 
    macierz_cell = macierz_cell - min_value_macierz_cell

    # Wyciągnięcie macierzy zawierającej maskę
    cell_area = image[:, 6, :, :]
    cell_area = cell_area/255

    # Maskowanie obszaru komórki
    macierz_cell = macierz_cell * cell_area
    return(macierz_cell, cell_area)

def znajdz_najlepszy_kwadrat(obraz):
    # Wymiary oryginalnego obrazu
    wysokosc = len(obraz)
    szerokosc = len(obraz[0])
    
    # Ustal długość boku kwadratu jako długość krótszego boku obrazu
    bok_kwadratu = min(wysokosc, szerokosc)
    
    # Zmienna pomocnicza do przechowywania najlepszej liczby jedynek oraz współrzędnych
    max_jedynek = 0
    najlepsze_wspolrzedne = (0, 0)  # domyślnie lewy górny róg
    
    # Przesuwamy kwadrat o boku `bok_kwadratu` przez cały obraz, aby znaleźć najlepszy obszar
    for i in range(wysokosc - bok_kwadratu + 1):
        for j in range(szerokosc - bok_kwadratu + 1):
            # Liczymy liczbę jedynek w bieżącym kwadracie o wymiarach bok_kwadratu x bok_kwadratu
            licznik_jedynek = sum(
                obraz[x][y] 
                for x in range(i, i + bok_kwadratu) 
                for y in range(j, j + bok_kwadratu)
            )
            
            # Aktualizujemy, jeśli znaleźliśmy większą liczbę jedynek
            if licznik_jedynek > max_jedynek:
                max_jedynek = licznik_jedynek
                najlepsze_wspolrzedne = (i, j)
    
    # Wycinamy najlepszy kwadrat z obrazu
    i_start, j_start = najlepsze_wspolrzedne
    przyciety_obraz = [
        row[j_start:j_start + bok_kwadratu]
        for row in obraz[i_start:i_start + bok_kwadratu]
    ]
    
    # Zwracamy współrzędne najlepszego kwadratu oraz przycięty obraz
    wspolrzedne_przycietego = (i_start, j_start, i_start + bok_kwadratu - 1, j_start + bok_kwadratu - 1)
    return wspolrzedne_przycietego, przyciety_obraz

# Selekcja kwadratu z komórką
def selsct_best_squere(cell_area):
    # Tworzymy tablicę Size_area i wypełniamy ją sumami dla każdego plastra
    Size_area = np.sum(cell_area, axis=(1, 2))

    # Znajdujemy indeks plastra o największej sumie pikseli
    max_index = np.argmax(Size_area)

    # Zapisujemy plaster o największej sumie do nowej zmiennej
    max_slice = cell_area[max_index]

    # Wywołanie funkcji
    wspolrzedne, przyciety_obraz = znajdz_najlepszy_kwadrat(max_slice)

    max_slice_squre = max_slice[wspolrzedne[0] : wspolrzedne[2], wspolrzedne[1] : wspolrzedne[3]]
    return wspolrzedne

# Tworzenie maski i procesing
def load_mask_EW():
    # Ścierzka do pliku z zakresem częstości OCT
    file_path = './EW_int_720_45.mat'

    # Wczytanie pliku
    mask = loadmat(file_path)

    # Dostęp do zmiennej EW_int (przykładowo, jeśli taka istnieje w pliku)
    EW_int = mask['EW_int']

    '''# Ścierzka do pliku z zakresem częstości OCT
    file_path = './New_V2_EW.mat'

    # Wczytanie pliku
    mask = loadmat(file_path)

    # Dostęp do zmiennej EW_int (przykładowo, jeśli taka istnieje w pliku)
    EW_bool = mask['EW']

    # Sprawdzenie kształtu macierzy
    print(EW_bool.shape)'''

    # Zamiana wymiarów na (180, 720, 720)
    mask_matrix = np.transpose(EW_int, (2, 0, 1))

    # Zapisz jako wielowarstwowy plik TIFF
    tifffile.imwrite('mask_matrix_small.tiff', mask_matrix)

    # Odczytanie obrazu 3D z pliku TIFF
    #mask_matrix = tifffile.imread('mask_matrix_small.tiff')

    # Sprawdzenie nowego kształtu macierzy
    #print("Rozmiar maski:", mask_matrix.shape)
    return mask_matrix

# Interpolacja Kubiczna
'''def rescale_to_target_shape(matrix, target_shape):
    # Obliczenie współczynników skalowania
    scale_factors = tuple(t / o for t, o in zip(target_shape, matrix.shape))
    
    # Przeskalowanie macierzy
    rescaled_matrix = zoom(matrix, scale_factors, order=3)  # 'order=3' dla interpolacji kubicznej
    
    return rescaled_matrix'''

# Interpolacja najbliższego sąsiada
def rescale_to_target_shape(matrix, target_shape):
    """
    Funkcja skaluje macierz do podanego rozmiaru, nie wprowadzając wartości ujemnych.
    Wykorzystuje interpolację najbliższego sąsiada (order=0) dla minimalnego zniekształcenia danych.
    
    Args:
    - matrix (np.ndarray): Macierz 3D do przeskalowania.
    - target_shape (tuple): Docelowy rozmiar macierzy.
    
    Returns:
    - np.ndarray: Przeskalowana macierz 3D.
    """
    # Obliczenie współczynników skalowania dla każdego wymiaru
    scale_factors = tuple(t / o for t, o in zip(target_shape, matrix.shape))
    
    # Przeskalowanie macierzy z użyciem interpolacji najbliższego sąsiada
    rescaled_matrix = zoom(matrix, scale_factors, order=0)
    
    return rescaled_matrix

# Wykonaj transformatę Fouriera na macierzy 3D
def cut_Fourier_spectrum(rescaled_matrix_cell, mask_matrix):
    start_time = time.time()

    fourier_transform = np.fft.fftn(rescaled_matrix_cell)

    # Przesunięcie zero częstotliwości do środka
    fourier_transform_shifted = np.fft.fftshift(fourier_transform)

    # Wycięcie widma charakterystyczneo dla OCT
    masked_fourier_transform = fourier_transform_shifted * mask_matrix

    '''viewer.add_image(np.log(1 + np.abs(fourier_transform_shifted)) , name="Widmo logarytmiczne", colormap='viridis')
    viewer.add_image(np.log(1 + np.abs(masked_fourier_transform)) , name="Wtcięte widmo", colormap='viridis')'''

    # Przeprowadzenie odwrotnej FFT
    ifft_image_shifted = ifftshift(masked_fourier_transform)  # Przesunięcie odwrotne
    inverse_fourier_transform = np.fft.ifftn(ifft_image_shifted)

    # Sprawdzenie kształtu macierzy
    #print("Po Fourierze:", inverse_fourier_transform.shape)

    # Wyciągnij część rzeczywistą
    real_part_inverse_fourier_transform = np.real(inverse_fourier_transform)

    # Zapis macierzy OCT
    #tifffile.imwrite('OCT_matrix.tiff', real_part_inverse_fourier_transform)

    # Zakończenie pomiaru czasu
    end_time = time.time()

    # Wyliczenie czasu trwania
    execution_time = end_time - start_time
    print(f"Czas wykonania FFT i IFFT: {execution_time} sekund")
    return real_part_inverse_fourier_transform

# Zapis do jednego pliku HDF5
def save_to_H5_file(file_name, rescaled_matrix_cell, real_part_inverse_fourier_transform):
    # Usunięcie końcówki '.ome.tif' i dodanie '_converted.h5'
    new_filename = file_name.replace(".ome.tif", "_converted.h5")
    with h5py.File(new_filename, 'w') as f:
        f.create_dataset('Cell', data = rescaled_matrix_cell) 
        f.create_dataset('OCT', data = real_part_inverse_fourier_transform)


# Path to the directory
data_directory = './'

# Wczytanie obrazów
pliki = sorted([f for f in os.listdir(data_directory) if f.endswith('.tif')])
images = [io.imread(os.path.join(data_directory, plik)) for plik in pliki]

cnt = 0
for i, file_name in enumerate(sorted(os.listdir(data_directory))):
    if file_name.endswith('.tif'):
        print("Nazwa pliku:", file_name, "  nr.:", cnt)
        image_path = os.path.join(data_directory, file_name)
        # Wczytanie macierzy
        image = io.imread(image_path)

        cell_matrix, cell_area = masking_cell_area(image)
        
        coordinates = selsct_best_squere(cell_area)

        cell_matrix = cell_matrix[:, coordinates[0] : coordinates[2], coordinates[1] : coordinates[3]]

        EW_mask = load_mask_EW()

        # Zmiana typu danych na int16
        cell_matrix = cell_matrix.astype(np.int16)
        print("Typ danych macierzy Cell",cell_matrix.dtype)

        # Przeskalowanie macierzy do wymiaru maski
        rescaled_matrix_cell = rescale_to_target_shape(cell_matrix, EW_mask.shape)

        # Symulacja OCT
        OCT_cell_matrix = cut_Fourier_spectrum(rescaled_matrix_cell, EW_mask)

        OCT_cell_matrix = OCT_cell_matrix.astype(np.float32)
        
        # Zapis do pliku HDF5
        save_to_H5_file(file_name, rescaled_matrix_cell, OCT_cell_matrix)
        
        cnt += 1

# Initialize the viewer
viewer = napari.Viewer()
viewer.add_image(cell_matrix , name="Cell", colormap='viridis')
viewer.add_image(OCT_cell_matrix , name="OCT", colormap='viridis')
print("Finished!")
# Start the napari viewer
napari.run()
```

```python
#Main 31.10.2024

import napari
from skimage import io
import os

import numpy as np
from scipy.io import savemat
from scipy.io import loadmat

from numpy.fft import fft2, ifft2, fftshift, ifftshift

from scipy.ndimage import median_filter

import tifffile

import h5py

# Path to the directory
data_directory = './'

# Initialize the viewer
viewer = napari.Viewer()

# Wczytanie obrazów
pliki = sorted([f for f in os.listdir(data_directory) if f.endswith('.tif')])
images = [io.imread(os.path.join(data_directory, plik)) for plik in pliki]

'''# Load images and add to viewer with different colors for each channel
colors = ['red', 'green', 'blue', 'magenta', 'cyan', 'yellow']'''
cnt = 0
for i, file_name in enumerate(sorted(os.listdir(data_directory))):
    if file_name.endswith('.tif'):
        print(file_name)
        image_path = os.path.join(data_directory, file_name)
        image = io.imread(image_path)
        name = "I_" + str(cnt)
        layer_number = "I_" + str(cnt)
        viewer.add_image(image, name=name)
        cnt += 1

liczba_wymiarow = image.shape
print("Liczba wymiarów macierzy:", liczba_wymiarow)

'''# Wybór jednej macierzy 3D
macierz_cell = image[:, 2, 216-126:216, :]

# Wybór jednej macierzy 3D
cell_area = image[:, 6, 216-126:216, :]
cell_area = cell_area/255
print(cell_area[21, :, 50])
min_value_macierz_cell = np.min(macierz_cell)

macierz_cell = macierz_cell - min_value_macierz_cell
print(macierz_cell[21, :, 50])'''

# Wyciągnięcie macierzy zawierającej komórkę
macierz_cell = image[:, 2, :, :]

#################################################################################################
# Sprawdzenie rozmiaru macierzy
rozmiar = macierz_cell.shape

# Obliczenie podstawowych statystyk
maksymalna_wartosc = np.max(macierz_cell)
minimalna_wartosc = np.min(macierz_cell)
srednia_wartosc = np.mean(macierz_cell)

# Wyświetlenie wyników
print("Informacje o macierzy macierz_celll:")
print("Rozmiar:", rozmiar)
print("Maksymalna wartość woksela:", maksymalna_wartosc)
print("Minimalna wartość woksela:", minimalna_wartosc)
print("Średnia wartość woksela:", srednia_wartosc)
print(macierz_cell[:, 109, 109])
#################################################################################################

# Normalizacja pikseli do 1
min_value_macierz_cell = np.min(macierz_cell) - 1 
macierz_cell = macierz_cell - min_value_macierz_cell

# Wyciągnięcie macierzy zawierającej maskę
cell_area = image[:, 6, :, :]
cell_area = cell_area/255

# Maskowanie obszaru komórki
macierz_cell = macierz_cell * cell_area

liczba_wymiarow = macierz_cell.shape
print("Liczba wymiarów macierzy:", liczba_wymiarow)

# Zapisz macierz do pliku .mat
savemat("matrix.mat", {"matrix_cell": macierz_cell})

#############################################################################
# Selekcja kwadratu z komórką

# Tworzymy tablicę Size_area i wypełniamy ją sumami dla każdego plastra
Size_area = np.sum(cell_area, axis=(1, 2))

# Znajdujemy indeks plastra o największej sumie pikseli
max_index = np.argmax(Size_area)

# Zapisujemy plaster o największej sumie do nowej zmiennej
max_slice = cell_area[max_index]

viewer.add_image(max_slice, name="Max")

'''def largest_possible_square(matrix):
    # Znajdujemy współrzędne pikseli o wartości 1
    one_positions = np.argwhere(matrix == 1)
    
    # Jeśli nie ma pikseli o wartości 1, zwracamy puste dane
    if one_positions.size == 0:
        return 0, (0, 0), np.array([])

    # Znajdź minimalne i maksymalne współrzędne pikseli 1
    min_row, min_col = one_positions.min(axis=0)
    max_row, max_col = one_positions.max(axis=0)
    
    # Obliczamy wymiary obejmujące wszystkie piksele 1
    height = max_row - min_row + 1
    width = max_col - min_col + 1
    side_length = max(height, width)  # Długość boku największego kwadratu
    
    # Ustalamy granice tak, by kwadrat obejmował wszystkie piksele 1
    row_start = max(0, min_row - (side_length - height) // 2)
    col_start = max(0, min_col - (side_length - width) // 2)
    
    # Sprawdzamy, czy kwadrat zmieści się w granicach macierzy
    row_start = min(row_start, matrix.shape[0] - side_length)
    col_start = min(col_start, matrix.shape[1] - side_length)

    # Wycinamy największy możliwy kwadrat zawierający wszystkie piksele 1
    largest_square_matrix = matrix[row_start:row_start + side_length, col_start:col_start + side_length]
    
    return side_length, (row_start, col_start), largest_square_matrix'''

def znajdz_najlepszy_kwadrat(obraz):
    # Wymiary oryginalnego obrazu
    wysokosc = len(obraz)
    szerokosc = len(obraz[0])
    
    # Ustal długość boku kwadratu jako długość krótszego boku obrazu
    bok_kwadratu = min(wysokosc, szerokosc)
    
    # Zmienna pomocnicza do przechowywania najlepszej liczby jedynek oraz współrzędnych
    max_jedynek = 0
    najlepsze_wspolrzedne = (0, 0)  # domyślnie lewy górny róg
    
    # Przesuwamy kwadrat o boku `bok_kwadratu` przez cały obraz, aby znaleźć najlepszy obszar
    for i in range(wysokosc - bok_kwadratu + 1):
        for j in range(szerokosc - bok_kwadratu + 1):
            # Liczymy liczbę jedynek w bieżącym kwadracie o wymiarach bok_kwadratu x bok_kwadratu
            licznik_jedynek = sum(
                obraz[x][y] 
                for x in range(i, i + bok_kwadratu) 
                for y in range(j, j + bok_kwadratu)
            )
            
            # Aktualizujemy, jeśli znaleźliśmy większą liczbę jedynek
            if licznik_jedynek > max_jedynek:
                max_jedynek = licznik_jedynek
                najlepsze_wspolrzedne = (i, j)
    
    # Wycinamy najlepszy kwadrat z obrazu
    i_start, j_start = najlepsze_wspolrzedne
    przyciety_obraz = [
        row[j_start:j_start + bok_kwadratu]
        for row in obraz[i_start:i_start + bok_kwadratu]
    ]
    
    # Zwracamy współrzędne najlepszego kwadratu oraz przycięty obraz
    wspolrzedne_przycietego = (i_start, j_start, i_start + bok_kwadratu - 1, j_start + bok_kwadratu - 1)
    return wspolrzedne_przycietego, przyciety_obraz

# Wywołanie funkcji
wspolrzedne, przyciety_obraz = znajdz_najlepszy_kwadrat(max_slice)
print("Współrzędne przyciętego obrazu:", wspolrzedne)
print("Współrzędna 0 przyciętego obrazu:", wspolrzedne[0])
print("Współrzędna 1 przyciętego obrazu:", wspolrzedne[1])
print("Współrzędna 2 przyciętego obrazu:", wspolrzedne[2])

max_slice_squre = max_slice[wspolrzedne[0] : wspolrzedne[2], wspolrzedne[1] : wspolrzedne[3]]

viewer.add_image(max_slice_squre, name="Largest_squere")

###################################################################################################
# Warning tu przycinamy największy kwadrat
###################################################################################################
macierz_cell = macierz_cell[:, wspolrzedne[0] : wspolrzedne[2], wspolrzedne[1] : wspolrzedne[3]]

viewer.add_image(macierz_cell, name="Cell")

#############################################################################
# Tworzenie maski i procesing

# Ścierzka do pliku z zakresem częstości OCT
file_path = './EW_int_528_66.mat'

# Wczytanie pliku
mask = loadmat(file_path)

# Dostęp do zmiennej EW_int (przykładowo, jeśli taka istnieje w pliku)
EW_int = mask['EW_int']

'''# Ścierzka do pliku z zakresem częstości OCT
file_path = './New_V2_EW.mat'

# Wczytanie pliku
mask = loadmat(file_path)

# Dostęp do zmiennej EW_int (przykładowo, jeśli taka istnieje w pliku)
EW_bool = mask['EW']

# Sprawdzenie kształtu macierzy
print(EW_bool.shape)'''

# Zamiana wymiarów na (180, 720, 720)
mask_matrix = np.transpose(EW_int, (2, 0, 1))

# Zapisz jako wielowarstwowy plik TIFF
tifffile.imwrite('mask_matrix_small.tiff', mask_matrix)

# Odczytanie obrazu 3D z pliku TIFF
#mask_matrix = tifffile.imread('mask_matrix_small.tiff')

# Sprawdzenie nowego kształtu macierzy
print("Rozmiar maski:", mask_matrix.shape)

viewer.add_image(mask_matrix , name="Mask", colormap='viridis')

###########################################################################
# Przeskalowanie macierzy do (180, 720, 720)

from scipy.ndimage import zoom

'''def rescale_to_target_shape(matrix, target_shape=(228, 912, 912)):
    # Obliczenie współczynników skalowania
    scale_factors = tuple(t / o for t, o in zip(target_shape, matrix.shape))
    
    # Przeskalowanie macierzy
    rescaled_matrix = zoom(matrix, scale_factors, order=3)  # 'order=3' dla interpolacji kubicznej
    
    return rescaled_matrix'''

def rescale_to_target_shape(matrix, target_shape):
    """
    Funkcja skaluje macierz do podanego rozmiaru, nie wprowadzając wartości ujemnych.
    Wykorzystuje interpolację najbliższego sąsiada (order=0) dla minimalnego zniekształcenia danych.
    
    Args:
    - matrix (np.ndarray): Macierz 3D do przeskalowania.
    - target_shape (tuple): Docelowy rozmiar macierzy.
    
    Returns:
    - np.ndarray: Przeskalowana macierz 3D.
    """
    # Obliczenie współczynników skalowania dla każdego wymiaru
    scale_factors = tuple(t / o for t, o in zip(target_shape, matrix.shape))
    
    # Przeskalowanie macierzy z użyciem interpolacji najbliższego sąsiada
    rescaled_matrix = zoom(matrix, scale_factors, order=0)
    
    return rescaled_matrix

print("Typ danych macierzy przed",macierz_cell.dtype)
macierz_cell = macierz_cell.astype(np.int16)
print("Typ danych macierzy po",macierz_cell.dtype)

# Przeskalowanie macierzy do wymiaru (180, 720, 720)
rescaled_matrix_cell = rescale_to_target_shape(macierz_cell, mask_matrix.shape)

print("Typ danych macierzy",rescaled_matrix_cell.dtype)

#tifffile.imwrite('cell_matrix.tiff', rescaled_matrix_cell)

viewer.add_image(rescaled_matrix_cell , name="Rescaled cell", colormap='viridis')

# Sprawdzenie nowego kształtu macierzy
print("Nowy kształt macierzy:", rescaled_matrix_cell.shape)

# Filtracja medianowa
'''rescaled_matrix_cell = median_filter(rescaled_matrix_cell, size=(3, 3, 3))

viewer.add_image(rescaled_matrix_cell , name="Medfilt_cell", colormap='viridis')'''

############################################################################
# Wykonaj transformatę Fouriera na macierzy 3D
import time
start_time = time.time()

fourier_transform = np.fft.fftn(rescaled_matrix_cell)

# Przesunięcie zero częstotliwości do środka
fourier_transform_shifted = np.fft.fftshift(fourier_transform)

'''# Wyciągnij część rzeczywistą
real_fourier_transform = np.real(fourier_transform)

# Wyciągnij część urojoną
imag_fourier_transform = np.imag(fourier_transform)

# Maska do wycięcia danych
masked_fourier_transform = real_fourier_transform * mask_matrix'''

masked_fourier_transform = fourier_transform_shifted * mask_matrix

viewer.add_image(np.log(1 + np.abs(fourier_transform_shifted)) , name="Widmo logarytmiczne", colormap='viridis')
viewer.add_image(np.log(1 + np.abs(masked_fourier_transform)) , name="Wtcięte widmo", colormap='viridis')

'''# Wykonaj odwrotną transformatę Fouriera
inverse_fourier_transform = np.fft.ifftn(masked_fourier_transform)'''

# Przeprowadzenie odwrotnej FFT
ifft_image_shifted = ifftshift(masked_fourier_transform)  # Przesunięcie odwrotne
inverse_fourier_transform = np.fft.ifftn(ifft_image_shifted)
#abs_inverse_fourier_transform = np.abs(inverse_fourier_transform)

# Sprawdzenie kształtu macierzy
print("Po Fourierze:", inverse_fourier_transform.shape)

# Wyciągnij część rzeczywistą
real_part_inverse_fourier_transform = np.real(inverse_fourier_transform)

# Zapis macierzy OCT
#tifffile.imwrite('OCT_matrix.tiff', real_part_inverse_fourier_transform)

# Zakończenie pomiaru czasu
end_time = time.time()

# Wyliczenie czasu trwania
execution_time = end_time - start_time
print(f"Czas wykonania: {execution_time} sekund")

#viewer.add_image(np.abs(inverse_fourier_transform) , name="abs inverse Fourier", colormap='viridis')
viewer.add_image(real_part_inverse_fourier_transform , name="Real part after Fourier", colormap='viridis')
########################################################################

# Sprawdzenie rozmiaru macierzy
rozmiar = macierz_cell.shape

# Obliczenie podstawowych statystyk
maksymalna_wartosc = np.max(macierz_cell)
minimalna_wartosc = np.min(macierz_cell)
srednia_wartosc = np.mean(macierz_cell)

# Wyświetlenie wyników
print("Informacje o macierzy macierz_celll:")
print("Rozmiar:", rozmiar)
print("Maksymalna wartość woksela:", maksymalna_wartosc)
print("Minimalna wartość woksela:", minimalna_wartosc)
print("Średnia wartość woksela:", srednia_wartosc)
print(macierz_cell[:, 109, 109])


# Sprawdzenie rozmiaru macierzy
rozmiar = rescaled_matrix_cell.shape

# Obliczenie podstawowych statystyk
maksymalna_wartosc = np.max(rescaled_matrix_cell)
minimalna_wartosc = np.min(rescaled_matrix_cell)
srednia_wartosc = np.mean(rescaled_matrix_cell)

# Wyświetlenie wyników
print("Informacje o macierzy rescaled_matrix_cell:")
print("Rozmiar:", rozmiar)
print("Maksymalna wartość woksela:", maksymalna_wartosc)
print("Minimalna wartość woksela:", minimalna_wartosc)
print("Średnia wartość woksela:", srednia_wartosc)
print(rescaled_matrix_cell[:, 360, 360])

# Sprawdzenie rozmiaru macierzy
rozmiar = real_part_inverse_fourier_transform.shape

# Obliczenie podstawowych statystyk
maksymalna_wartosc = np.max(real_part_inverse_fourier_transform)
minimalna_wartosc = np.min(real_part_inverse_fourier_transform)
srednia_wartosc = np.mean(real_part_inverse_fourier_transform)

# Wyświetlenie wyników
print("Informacje o macierzy real_part_inverse_fourier_transform:")
print("Rozmiar:", rozmiar)
print("Maksymalna wartość woksela:", maksymalna_wartosc)
print("Minimalna wartość woksela:", minimalna_wartosc)
print("Średnia wartość woksela:", srednia_wartosc)
print(real_part_inverse_fourier_transform[:, 360, 360])

# Zapis do jednego pliku HDF5
with h5py.File('macierze.h5', 'w') as f:
    f.create_dataset('m1', data = rescaled_matrix_cell) 
    f.create_dataset('m2', data = real_part_inverse_fourier_transform)


real_part_inverse_fourier_transform = real_part_inverse_fourier_transform.astype(np.float32)
# Zapis do jednego pliku HDF5
with h5py.File('macierze_low.h5', 'w') as f:
    f.create_dataset('m1', data = rescaled_matrix_cell) 
    f.create_dataset('m2', data = real_part_inverse_fourier_transform)

# Zapisanie jako OME-TIFF
tifffile.imwrite('macierz.ome.tiff', real_part_inverse_fourier_transform, imagej=True)  # imagej=True dla poprawnego formatowania OME

print("Finished!")
# Start the napari viewer
napari.run()
```
